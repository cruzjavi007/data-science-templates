## ROC CURVE

```python
fig_roc, ax_roc = plt.subplots(figsize = (10.8, 10.8))
np.random.seed(0)

ROC_CURVE_VALUE = 5
ROC_CURVE_LABELS = ["Name" + str(i) for i in range(ROC_CURVE_VALUE)]

y_true_rc = [[int(np.random.uniform(0, 2)) for _ in range(MAX_VALUE ** 2)] for _ in range(ROC_CURVE_VALUE)]
y_pred_rc = [[int(np.random.uniform(0, 2)) for _ in range(MAX_VALUE ** 2)] for _ in range(ROC_CURVE_VALUE)]

for i in range(ROC_CURVE_VALUE):
    y_true = [label[i] for label in y_true_rc]
    y_pred = [label[i] for label in y_pred_rc]
    fpr, tpr, _ = roc_curve(y_true, y_pred)
    plt.plot(fpr, tpr, label = "{} - AUC: {:0.2f}".format(ROC_CURVE_LABELS[i], auc(fpr, tpr)))

plt.plot([0, 1], [0, 1], color = "darkblue", linestyle = "--", label = "Random Classifier - AUC: 0.5")
plt.axis("square")

plt.xticks(np.arange(0, 1.1, 0.1))
plt.yticks(np.arange(0, 1.1, 0.1))
plt.xlabel("False Positive Rate", fontsize = 20, labelpad = 10)
plt.ylabel("True Positive Rate", fontsize = 20, labelpad = 10)
plt.legend(fontsize = "10", loc = "lower right", title = "Area Under Curve (AUC)")

plt.show()
# fig_roc.savefig("ROC CURVE.png")
```

## REGRESSION

```python
fig_regression1, ax_regression1 = plt.subplots(figsize = (10.8, 10.8))
np.random.seed(0)

REGRESSION1_VALUE = 100
regression1_values_true = [np.random.rand() for _ in range(REGRESSION1_VALUE)]
regression1_values_pred = [np.random.rand() for _ in range(REGRESSION1_VALUE)]
regression1_values = pd.DataFrame({"true": regression1_values_true, "pred": regression1_values_pred})

sb.kdeplot(data = regression1_values,
           x = "true",
           y = "pred",
           fill = True,
           levels = 25,
           cmap = "rocket_r",
           ax = ax_regression1)

plt.xlabel("Predicted Value", fontsize = 20, labelpad = 10)
plt.ylabel("True Value", fontsize = 20, labelpad = 10)

plt.show()
# fig_regression1.savefig("REGRESSION1.png")
```

```python
fig_regression2, ax_regression2 = plt.subplots(figsize = (10.8, 10.8))
np.random.seed(0)

REGRESSION2_VALUE = 100
regression2_values_true = [np.random.rand() for _ in range(REGRESSION2_VALUE)]
regression2_values_pred = [np.random.rand() for _ in range(REGRESSION2_VALUE)]
regression2_values = pd.DataFrame({"true": regression2_values_true, "pred": regression2_values_pred})

regression2_values_min_score = min(min(regression2_values["true"]), min(regression2_values["pred"]))
regression2_values_max_score = max(max(regression2_values["true"]), max(regression2_values["pred"]))

sb.scatterplot(data = regression2_values,
               x = "true",
               y = "pred",
               color = "orange",
               ax = ax_regression2)

ax_regression2.plot([regression2_values_min_score, regression2_values_max_score],
                    [regression2_values_min_score, regression2_values_max_score],
                    "k--")

plt.xticks(np.arange(0, 1.1, 0.1))
plt.yticks(np.arange(0, 1.1, 0.1))
plt.xlabel("Predicted Value", fontsize = 20, labelpad = 10)
plt.ylabel("True Value", fontsize = 20, labelpad = 10)

plt.show()
# fig_regression2.savefig("REGRESSION2.png")
```
